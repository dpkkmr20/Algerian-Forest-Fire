```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Updated_Algerian_Forest_Fires.csv")
df
```

```python
#drop date month and year
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## Independent And dependent features
x=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
x.head()
```

```python
y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlaltion
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
X_train.corr()
```

```python
def correlation(dataset, threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        #print("Column : ",corr_matrix.columns[i],i)
        for j in range(i):
             if abs(corr_matrix.iloc[i, j]) > threshold:
                #print(corr_matrix.iloc[i, j])
                colname = corr_matrix.columns[i]
                #print(colname)
                col_corr.add(colname)
             #print(corr_matrix.columns[j],j)
        #print(" ")
    return col_corr
    
```

```python
corr_features=correlation(X_train,0.85)
```

```python
corr_features
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

## Feature Scaling or standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

## Box plot to understand effect of standardization

```python
plt.subplots(figsize=(15,20))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_train_Before_Scaling")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X_train after scaling")
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
linear_reg=LinearRegression()
linear_reg.fit(X_train_scaled,y_train)
y_pred=linear_reg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mean absolute error : ",mae)
print("R2 score : ",score)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error,r2_score
lasso_reg=Lasso()
lasso_reg.fit(X_train_scaled,y_train)
y_pred=lasso_reg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mean absolute error : ",mae)
print("R2 score : ",score)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error,r2_score
ridge_reg=Ridge()
ridge_reg.fit(X_train_scaled,y_train)
y_pred=ridge_reg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mean absolute error : ",mae)
print("R2 score : ",score)
```

## Elastic net

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error,r2_score
en_reg=ElasticNet()
en_reg.fit(X_train_scaled,y_train)
y_pred=en_reg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mean absolute error : ",mae)
print("R2 score : ",score)
```
